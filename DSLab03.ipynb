```python
df['Family_size'] = df['SibSp'] + df['Parch'] +1
```

```python
# family_type
# 1 -> alone
# 2-4 -> small
# >5 -> large

def transform_family_size(num):
    if num == 1:
        return 'alone'
    elif num > 1 and num <5:
        return 'small'
    else:
        return 'large'
```

```python
df['family_type'] = df['Family_size'].apply(transform_family_size)
```
