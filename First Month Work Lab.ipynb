```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
df.columns
```

```python
# Exploratory Data Analysis
# plotting a bar chart for Gender and it's count
ax = sns.countplot(x = 'Gender', data = df)+
sns.set(rc={'figure.figsize':(5,4)})
```

```python
ax = sns.countplot(x = 'Gender',data = df)
sns.set(rc={'figure.figsize':(7,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total income
gen_income = df.groupby(['Gender'], as_index = False)['Income'].sum().sort_values(by='Income', ascending = False)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x = 'Gender', y = 'Income', data = gen_income)
```

```python
# Group by Gender and calculate total income and count of each gender
gen_income = df.groupby('Gender').agg({'Income': 'sum', 'Gender': 'count'}).rename(columns={'Gender': 'Count'}).reset_index()

# Plotting total income
sns.set(rc={'figure.figsize': (14, 5)})

fig, ax1 = plt.subplots()

# Bar plot for total income
sns.barplot(x='Gender', y='Income', data=gen_income, ax=ax1, palette='viridis')
ax1.set_ylabel('Total Income', fontsize=14)

# Creating a second y-axis for the count of each gender
ax2 = ax1.twinx()
sns.barplot(x='Gender', y='Count', data=gen_income, ax=ax2, alpha=0.3, palette='pastel')
ax2.set_ylabel('Count', fontsize=14)

# Setting the title
plt.title('Total Income and Count by Gender', fontsize=16)

# Show the plot
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'Age', hue = 'Gender')

sns.set(rc={'figure.figsize':(25,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Income vs Age
sales_age = df.groupby(['Age'], as_index=False)['Income'].sum().sort_values(by='Income', ascending=True)
sns.set(rc={'figure.figsize':(15,5)})

sns.barplot(x = 'Age',y= 'Income' ,data = sales_age)
```

```python

# Group by Age and calculate the maximum income for each age
sales_age = df.groupby('Age', as_index=False)['Income'].max().sort_values(by='Income', ascending=True)

# Set figure size
sns.set(rc={'figure.figsize': (35, 20)})

# Create the bar plot
ax = sns.barplot(x='Age', y='Income', data=sales_age, palette='viridis')

# Add bar labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')

# Set the title and labels
plt.title('Highest Income by Age', fontsize=50)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Income', fontsize=14)

# Show the plot
plt.show()
```

```python
# Bin the ages into three groups
bins = [0, 30, 50, 100]
labels = ['0-30', '31-50', '51+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Create the countplot
sns.set(rc={'figure.figsize': (25, 10)})
ax = sns.countplot(data=df, x='AgeGroup', hue='Gender')

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])

# total number of Gender from top 10 DiningOccasion

gen_DO = df.groupby(['DiningOccasion'], as_index=False)['Gender'].sum().sort_values(by='Gender', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

gen_DO = sns.barplot(data=df, x='DiningOccasion', y='Gender')

# Create a count plot to show separate counts for Male and Female
gen_DO = sns.countplot(data=df, x='DiningOccasion', hue='Gender')


# Get the containers from the plot
containers = gen_DO.containers

# Add bar labels
for bars in containers:
    gen_DO.bar_label(bars)
```

```python
# total number of Gender from top 10 VisitFrequency

gen_DO = df.groupby(['VisitFrequency'], as_index=False)['Gender'].sum().sort_values(by='Gender', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

gen_DO = sns.barplot(data=df, x='VisitFrequency', y='Gender')

# Create a count plot to show separate counts for Male and Female
gen_DO = sns.countplot(data=df, x='VisitFrequency', hue='Gender')


# Get the containers from the plot
containers = gen_DO.containers

# Add bar labels
for bars in containers:
    gen_DO.bar_label(bars)
```

```python
# total number of Gender from top 10 TimeOfVisit

gen_DO = df.groupby(['TimeOfVisit'], as_index=False)['Gender'].sum().sort_values(by='Gender', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

gen_DO = sns.barplot(data=df, x='TimeOfVisit', y='Gender')

# Create a count plot to show separate counts for Male and Female
gen_DO = sns.countplot(data=df, x='TimeOfVisit', hue='Gender')


# Get the containers from the plot
containers = gen_DO.containers

# Add bar labels
for bars in containers:
    gen_DO.bar_label(bars)
```

```python
# total number of Gender from top 10 PreferredCuisine

gen_DO = df.groupby(['PreferredCuisine'], as_index=False)['Gender'].sum().sort_values(by='Gender', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

gen_DO = sns.barplot(data=df, x='PreferredCuisine', y='Gender')

# Create a count plot to show separate counts for Male and Female
gen_DO = sns.countplot(data=df, x='PreferredCuisine', hue='Gender')


# Get the containers from the plot
containers = gen_DO.containers

# Add bar labels
for bars in containers:
    gen_DO.bar_label(bars)
```

```python
# Group by 'PreferredCuisine' and aggregate 'Gender' and 'FoodRating'
agg_df = df.groupby(['PreferredCuisine'], as_index=False).agg({'Gender': 'sum', 'FoodRating': 'mean'})

# Sort values and get top 10 'PreferredCuisine'
top_10 = agg_df.sort_values(by='Gender', ascending=False).head(10)

# Plotting
sns.set(rc={'figure.figsize': (15, 5)})

# Bar plot for 'Gender'
gen_DO = sns.barplot(data=top_10, x='PreferredCuisine', y='Gender', palette='viridis')

# Adding bar labels
for container in gen_DO.containers:
    gen_DO.bar_label(container)

# Adding FoodRating to the plot as text annotations
for index, row in top_10.iterrows():
    gen_DO.text(index, row['Gender'], round(row['FoodRating'], 2), color='black', ha="center")

plt.title('Gender Distribution and Food Rating by Preferred Cuisine')
plt.xlabel('Preferred Cuisine')
plt.ylabel('Total Gender Count')
plt.show()
```
